## First graph csv: Dot Density (one symbol, one color)

To make the Graph more readable we count the tree with same scientific name and common name that are near each other.

In this way we will have instead of having 13 raws with the same tree all in the same place just few meters away from each other, we'll have just a raw of value 13.

```python
latitude_coordinate = Y_state_coor['latitude_coordinate']
longitude_coordinate = Y_state_coor['longitude_coordinate']
longitude_coordinate=longitude_coordinate.round(1)
latitude_coordinate=latitude_coordinate.round(1)
```

```python
Y_state_coor['long_coor'] = longitude_coordinate
Y_state_coor['lat_coor'] = latitude_coordinate
```

```python
count_treeNear=pd.DataFrame(Y_state_coor.groupby(['state','city', 'scientific_name',	'common_name','long_coor','lat_coor'])['lat_coor'].count().reset_index(name="value"))
count_treeNear
```

```python
# Merge DataFrames based on specified columns
merged_df = pd.merge(count_treeNear, Y_state_coor, on=['state', 'city', 'scientific_name', 'common_name', 'long_coor', 'lat_coor'], how='left')

# Update 'longitude_coordinate' and 'latitude_coordinate' columns in count_treeNear
count_treeNear['longitude_coordinate'] = merged_df['longitude_coordinate']
count_treeNear['latitude_coordinate'] = merged_df['latitude_coordinate']
```

```python
merged_df
```

```python
# Drop duplicates based on specified columns
result_df = merged_df.drop_duplicates(subset=['state', 'city', 'scientific_name', 'common_name', 'long_coor', 'lat_coor', 'value'])
result_df
```

```python
result_df2 = result_df[['state',	'city',	'scientific_name',	'common_name',	'value',	'longitude_coordinate',	'latitude_coordinate']]
```

## Third graph: Aboundance

Here we simply count tree per each State

```python
count=pd.DataFrame(Y_state_coor.groupby(['state','city','long_coor','lat_coor'])['lat_coor'].count().reset_index(name="value"))
```

```python
# Group by 'state' and sum up the 'value' elements
summed_count = count.groupby('state')['value'].sum().reset_index()

# Display the result
summed_count
```

## Fourth graph: Density (per State)

```python
# Data
state_names = ['Wisconsin', 'California', 'Pennsylvania', 'Hawaii', 'Missouri',
               'District of Columbia', 'Ohio', 'Maryland', 'South Dakota',
               'Colorado', 'Iowa', 'Florida', 'Texas', 'Kansas', 'New York',
               'North Carolina', 'Kentucky', 'Washington', 'Tennessee', 'Nevada',
               'Michigan', 'New Mexico', 'Oklahoma', 'Oregon']

areas_sqkm = [140269, 423970, 116056, 16637, 178040,
              179, 116098, 25142, 196350,
              269601, 144669, 138887, 678058, 211753, 141297,
              126161, 102269, 184662, 109150, 286352,
              150504, 314917, 181037, 254799]

# Create DataFrame
area_df = pd.DataFrame({'State': state_names, 'Area_sqkm': areas_sqkm})
```

```python
# Merge the 'summed_count' and 'area_df' DataFrames on the 'State' column
merged_df2 = pd.merge(summed_count, area_df, how='inner', left_on='state', right_on='State')

# Calculate tree density (trees per square kilometer)
merged_df2['TreeDensity'] = merged_df2['value'] / merged_df2['Area_sqkm']

merged_df2=merged_df2[['state', 'TreeDensity', 'value', 'Area_sqkm']]
```

```python
density=merged_df2['TreeDensity'].round(2)
merged_df2['Density'] = density
merged_df2=merged_df2[['state', 'Density', 'value', 'Area_sqkm']]
merged_df2
```
