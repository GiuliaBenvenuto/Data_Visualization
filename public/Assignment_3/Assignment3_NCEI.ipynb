For the third Assignment: we use the data from the National Centers for Enviromental Information. Just like the second Assignment we decided to visualize the states by grouping them (West, South, Central Region...). To visualize better the change of the temperature over the years, we selected the years by decades, beginning with 1900 until 2020. We converted the temperatures from Farehnheit to Celcius. In the end for each region we computed the absolute minimum temperature and maximum temperature for each region to scale better the charts.

## Let's change the DataFrame format putting the years as the name of the columns..

To make our job easier we sorted our data in a dictionary... Then we save our result back in a Dataframe file

First, let's declare the table of the Months

```python
Months=['Jan',	'Feb',	'Mar',	'Apr',	'May',	'Jun',	'Jul',	'Aug',	'Sep',	'Oct',	'Nov',	'Dec']
```

```python
Months_df = pd.DataFrame(Months, columns=['Months'])
```

### MAX

```python
df_reg=dict()
for r in Max_Temp['region'].unique():
  temp=Max_Temp[Max_Temp['region']==r]
  df_reg[r]=temp
```

```python
Dict_Reg2=dict()
for dk in df_reg.keys():
  tmp_df=df_reg[dk]

  Dict_Year=dict()
  for y in Max_Temp['year'].unique():
    temp=tmp_df[tmp_df['year']==y]

    list_m=[]
    for m in Months:

      list_m.append((m, float(temp[m])))

    Dict_Year[y]=pd.DataFrame(list_m, columns=['Months', str(y)])

  Dict_Reg2[dk]=Dict_Year
```

## Absolute Min and Max of every region & Final CSV files

### Max temperature of each region:

```python
## EXTRACT ALL
def extract_selected_Y(dict_REG, years):
  #d=Dict_Reg2['Northwest Region']
  first=True
  for y in years:

    if first:
      df=dict_REG[y]
      first=False

    else:
      df=df.merge(dict_REG[y], on='Months')

  return df
```

For the final csv uncomment the last line

```python
years=[1900, 1910, 1920, 1930, 1940, 1950, 1960, 1970, 1980, 1990, 2000, 2010, 2020]
for d in Dict_Reg2.keys():
  d_csv=extract_selected_Y(Dict_Reg2[d], years)
  print(d)
  filename=d+'.csv'
  d_csv_max_index=pd.DataFrame(pd.DataFrame(d_csv.max(axis='index'), index=None)[0])
  list_csv=d_csv.max(axis='index').tail(-1).tolist()
  print(max(list_csv))
  #d_csv.to_csv(filename, index=False)
```
