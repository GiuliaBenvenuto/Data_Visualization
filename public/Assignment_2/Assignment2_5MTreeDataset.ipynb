## API request

Before sending request to the api let's clean the dataset from null and differently named scientific name

```python
Y_city_sciencename=pd.DataFrame(data[[ 'state', 'city','location_type', 'scientific_name']])
Y_city_sciencename=Y_city_sciencename.dropna()
Y_city_sciencename = Y_city_sciencename[Y_city_sciencename['location_type'] != 'no_info']
Y_city_sciencename = Y_city_sciencename[Y_city_sciencename['location_type'] != '<null>']
Y_city_sciencename = Y_city_sciencename[Y_city_sciencename['scientific_name'] != 'Prunus L.']
Y_city_sciencename = Y_city_sciencename[Y_city_sciencename['scientific_name'] != 'Malus Mill.']
Y_city_sciencename = Y_city_sciencename[Y_city_sciencename['scientific_name'] != 'Schinus terebinthifolius']
Y_city_sciencename = Y_city_sciencename[Y_city_sciencename['scientific_name'] != 'Mlaus scheideckeri']
Y_city_sciencename = Y_city_sciencename[Y_city_sciencename['scientific_name'] != 'Carya Nutt.']
Y_city_sciencename = Y_city_sciencename[Y_city_sciencename['scientific_name'] != 'Pterostyrax hispida']
Y_city_sciencename = Y_city_sciencename[Y_city_sciencename['scientific_name'] != 'Acer L.']
Y_city_sciencename = Y_city_sciencename[Y_city_sciencename['scientific_name'] != 'Viburnum L.']
Y_city_sciencename = Y_city_sciencename[Y_city_sciencename['scientific_name'] != 'Aesculus L.']
Y_city_sciencename = Y_city_sciencename[Y_city_sciencename['scientific_name'] != 'ÃƒÆ’Ã†â€™ÃƒÂ¢Ã¢â€šÂ¬Ã¢â‚¬ÂSycoparrotia P.']
Y_city_sciencename = Y_city_sciencename[Y_city_sciencename['scientific_name'] != 'Pyrus L.']
Y_city_sciencename = Y_city_sciencename[Y_city_sciencename['scientific_name'] != 'Populus L.']
Y_city_sciencename = Y_city_sciencename[Y_city_sciencename['scientific_name'] != "Acer platanoides 'Fairview'"]
Y_city_sciencename = Y_city_sciencename[Y_city_sciencename['scientific_name'] != "Roseodendron donnell-smithii"]
Y_city_sciencename = Y_city_sciencename[Y_city_sciencename['scientific_name'] != 'Prunnus yedoensis']
Y_city_sciencename = Y_city_sciencename[Y_city_sciencename['scientific_name'] != 'Gingko biloba']
Y_city_sciencename = Y_city_sciencename[Y_city_sciencename['scientific_name'] != 'Tilia sp.']
Y_city_sciencename = Y_city_sciencename[Y_city_sciencename['scientific_name'] != 'Ulmus hybrid']
```

```python
scient_name=Y_city_sciencename['scientific_name']
namelist=scient_name.drop_duplicates().to_list()
```

```python
compl_list=[]
for i in namelist:
  x = i.split()
  temp_list=[]
  for i_x in x:
    temp_list.append(i_x.lower())
  compl_list.append(temp_list)
```

Now api requests..

```python
data = {'scientific_name': [],
        'family': []}
df = pd.DataFrame(data)


for i in compl_list:
    if len(i) > 1:
      # Query URL
      url = ('https://api.gbif.org/v1'
            '/species?'
            'name='+ str(i[0]) +'%20'+ str(i[1]) +''  # Unit: CLV (2005) Million EUR
            )
      response = requests.get(url)  # Make a GET request to the URL
      print(f"{str(i[0])} {str(i[1])}: Request returned {response.status_code} : '{response.reason}'")
      payload = response.json()  # Parse `response.text` into JSON
      json_tree = payload['results']
      print(json_tree[0]['family'])
      new_row = {'scientific_name': ''+ str(i[0]) +' '+ str(i[1]) +'', 'family': ''+ str(json_tree[0]['family']) + ''}
      print(new_row)
      df.loc[len(df)] = new_row
    else:
      # Query URL
      url = ('https://api.gbif.org/v1'
            '/species?'
            'name='+ str(i[0]) +''  # Unit: CLV (2005) Million EUR
            )
      response = requests.get(url)  # Make a GET request to the URL
      print(f"{str(i[0])} : Request returned {response.status_code} : '{response.reason}'")
      payload = response.json()  # Parse `response.text` into JSON
      json_tree = payload['results']
      print(json_tree[0]['family'])
      new_row = {'scientific_name': ''+ str(i[0]) + '', 'family': ''+ str(json_tree[0]['family']) + ''}
      print(new_row)
      df.loc[len(df)] = new_row
```
